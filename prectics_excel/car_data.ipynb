```python
import pandas as pd
import numpy as np
df=pd.read_csv('used_cars_data.csv')
print(df.head())
```

```python
# drop column new_price
df=df.drop(columns=['New_Price'])
```

```python
print(df.isnull().sum())
```

```python
df['Price']=df['Price'].fillna(df['Price'].mean())
df['Seats']=df['Seats'].fillna(df['Seats'].mean())
print(df.isnull().sum())
```

```python
df.dropna(axis=0,inplace=True)
print(df.isnull().sum())
```

```python
def str_to_num(data):
    if isinstance(data,str):
        data=data.split()[0]
    return float(data)
df['Engine']=df['Engine'].apply(str_to_num)
df['Mileage']=df['Mileage'].apply(str_to_num)  
print(df.info())  
```

```python
print(df['Power'])
def clean(data):
    if isinstance (data,str):
        if "null" in data:
            data=float(0)
        else:
            data=data.split()[0]
    return float(data)
df['Power']=df['Power'].apply(clean)
```

```python
print(df['Power'])
df.to_csv('car_clean_dataset')
```

```python
group_Fuel_Type=df.groupby(['Fuel_Type'])['Fuel_Type'].count()
print(group_Fuel_Type)
```

```python
group_transmission=df.groupby(['Transmission'])['Transmission'].count()
print(group_transmission)
```

```python
group_Owner_Type=df.groupby(['Owner_Type'])['Owner_Type'].count()
print(group_Owner_Type)
```

```python
group_car=df.groupby(['Name'])['Name'].count()
group_car_sort=group_car.sort_values(ascending=False).head(20)
print(group_car_sort)
```

```python
for i in group_car_sort.index:
    index_value=df[df['Name']==i].index[0]
    print(df['Price'].iloc[index_value])
```
